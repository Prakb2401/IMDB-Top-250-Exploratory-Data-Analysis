```python
import requests
import pandas as pd
from api_key import OMBd_api
import random
import time
import json
from pprint import pprint
# Replace 'YOUR_API_KEY' with your actual OMDB API key
API_KEY = OMBd_api

# Function to fetch movie data by title
def fetch_movie_data(title):
    url = f'http://www.omdbapi.com/?apikey={OMBd_api}&t={title}'
    response = requests.get(url)
    data = response.json()
    return data

df_imdb = pd.read_csv('movies.csv')

# Sort the dataset by IMDb rating in descending order to get top-rated movies
df_imdb.sort_values(by='rank', ascending=False, inplace=True)

# Get the top 1000 movie titles from the dataset
top_1000_titles = df_imdb['name'][:1000]

# Create an empty list to store movie data
movie_data = []

# Fetch data for each movie title and add to the list
for title in top_1000_titles:
    data = fetch_movie_data(title)
    if data.get('Response') == 'True':
        movie_data.append(data)
        print(f"updated")
    else:
        print(f"Error fetching data for '{title}'")

    time.sleep(1)  # Pause 1 second to avoid overloading the API

# Create a DataFrame from the movie_data list
df = pd.DataFrame(movie_data)

# Display the DataFrame
pprint(df)
```
